```python
#############################
# Initilize 
#############################
# Import packages
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize


# Set global plot params
fontsize = 8
figsize = (6,4)
dpi = 300

plt.rcParams.update({'font.size': fontsize, 'figure.figsize': figsize, 'figure.dpi': dpi})
plt.rcParams['xtick.labelsize'] = fontsize
plt.rcParams['ytick.labelsize'] = fontsize
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
plt.rcParams['xtick.major.size'] = 4
plt.rcParams['ytick.major.size'] = 4
plt.rcParams['xtick.major.width'] = 0.5
plt.rcParams['ytick.major.width'] = 0.5



plt.rcParams['xtick.top'] = True
plt.rcParams['ytick.right'] = True
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
```

## ODEINT


```python
t,x,v = np.loadtxt('outputs/odeint.txt',unpack = True, skiprows = 1)
```

```python
# Plot position vs. time:
plt.plot(t,v[0]*np.sin(t),'r',label = 'Analytic Result', linewidth = 0.4)
plt.scatter(t,x, label = 'ODEINT')

plt.plot(t, x-v[0]*np.sin(t), label = 'Residual', color = 'k', linestyle = '-.', linewidth = 0.4)

plt.title(r'$\frac{d^2x}{dt^2} = -\omega^2 x$' + " [ODEINT]")
plt.xlabel("Time (s)")
plt.ylabel("Position (m)")
plt.legend(loc ='upper right', framealpha = 1, fancybox = 0, fontsize = 6)
```

```python
# caclulate energy:
# E = T+ V_sp
E = (0.5*v**2)+(0.5*x**2)

# Plot energy vs time:

plt.scatter(t,E,label = "ODEINT")
plt.title(r'$\frac{d^2x}{dt^2} = -\omega^2 x$' + " [ODEINT]")
plt.xlabel("Time (s)")
plt.ylabel("Energy (J)")
plt.ylim(0.499,0.501)

```

## Verlet Algorithm!

```python
# Read in c++ results
t,x,v,E = np.loadtxt('outputs/verlet.txt',unpack = True, skiprows = 1)
```

```python
# Plot position vs. time:
t_plot = np.linspace(0,np.max(t),500)
plt.plot(t_plot,v[0]*np.sin(t_plot),'r',label = 'Analytic Result', linewidth = 0.4)
plt.scatter(t,x, label = 'Verlet')
plt.plot(t, x-v[0]*np.sin(t), label = 'Residual', color = 'k', linestyle = '-.', linewidth = 0.4)

plt.title(r'$\frac{d^2x}{dt^2} = -\omega^2 x$' + " [Verlet Integration]")
plt.xlabel("Time (s)")
plt.ylabel("Position (m)")
plt.legend(loc ='upper right', framealpha = 1, fancybox = 0, fontsize = 6)
```

```python
plt.scatter(t,E)
plt.title(r'$\frac{d^2x}{dt^2} = -\omega^2 x$' + " [Velocity-Verlet Integration]")
plt.xlabel("Time (s)")
plt.ylabel("Energy(J)")

plt.ylim(0.4,0.6)
```
