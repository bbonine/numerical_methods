```python
#############################
# Initilize 
#############################
# Import packages
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize


# Set global plot params
fontsize = 8
figsize = (6,4)
dpi = 300

plt.rcParams.update({'font.size': fontsize, 'figure.figsize': figsize, 'figure.dpi': dpi})
plt.rcParams['xtick.labelsize'] = fontsize
plt.rcParams['ytick.labelsize'] = fontsize
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
plt.rcParams['xtick.major.size'] = 4
plt.rcParams['ytick.major.size'] = 4
plt.rcParams['xtick.major.width'] = 0.5
plt.rcParams['ytick.major.width'] = 0.5



plt.rcParams['xtick.top'] = True
plt.rcParams['ytick.right'] = True
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
```

```python
# Set up step info
steps = 100
x_range = 20*np.pi
h = x_range / steps
print(h)
```

```python
# Read in c++ results
c_xvals, euler_yvals, mod_euler_yvals = np.loadtxt('outputs/output_euler.txt', unpack = True)
```

```python
# Plot


#  analytical result: y(x) = -sinx
plt.figure(figsize = [8,6])
x = np.linspace(0,x_range, 1000)
plt.plot(x,-np.sin(x), label = 'Analytical Result', linewidth = 0.8, color = 'limegreen')
# Euler method:
plt.plot(c_xvals,euler_yvals, label = 'Euler Method', linewidth = 0.8, color = 'deeppink')
plt.plot(c_xvals,-np.sin(c_xvals)-euler_yvals, label = 'Residual', linestyle = '-.', c = 'k', linewidth = 0.4)

plt.title("Euler Method: " r'$\frac{df(x)}{dx} = -cos(x)$')
plt.ylim(-1.4,1.4)
plt.legend(fontsize = 6)


plt.xlabel('x')
plt.ylabel('f(x)')


plt.ylim(-1.4,1.4)

```

```python
# Plot the Euler-Predictor Method
#  analytical result: y(x) = -sinx
plt.figure(figsize = [8,6])

# Modified method:
plt.plot(c_xvals,mod_euler_yvals, label = 'Euler-Predictor Method', linewidth = 1.8, color = 'red')
x = np.linspace(0,x_range, steps)
plt.plot(x,-np.sin(x), label = 'Analytical Result', linewidth = 0.8, color = 'blue')
plt.plot(c_xvals,-np.sin(c_xvals)-mod_euler_yvals, label = 'Residual', linestyle = '-.', c = 'k', linewidth = 0.5)

plt.title("Modified-Euler Method: " r'$\frac{df(x)}{dx} = -cos(x)$')
plt.ylim(-1.4,1.4)
plt.legend(fontsize = 6)


plt.xlabel('x')
plt.ylabel('f(x)')


plt.ylim(-1.4,1.4)
```
